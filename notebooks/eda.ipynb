## Window Size Decision
For the Apple Stock dataset, each row is a day so a suitable window size 30 which is a month. Financial data has slower patterns, trends can take over weeks to form and market behavior is heavily influence by time. Too small window like 3 days or a week is useless because there are not enough context for meaningful prediction. Too large window like 100 200 days make the task harder to learn because of the high complexity, training process will be exceedingly slow.

For the Weather dataset, a row represent a datapoint every 10 mins, a suitable window size is 144 which is a full 24 hours day. Temperature have a daily cycle (domain knowledge). 

```python
subset = weather.iloc[1000:2000] 

plt.figure(figsize=(10,4))
plt.plot(subset["date"], subset["T"])
plt.title("Zoomed Temperature (Few Days)")
plt.show()
```

```python
weather["hour"] = weather["date"].dt.hour

hourly_avg = weather.groupby("hour")["T"].mean()

plt.plot(hourly_avg.index, hourly_avg.values)
plt.title("Average Temperature by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Temperature")
plt.show()
#confirmed the daily periodicity
```

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(weather["T"])
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(weather["T"], lags=300)
plt.show()
```

The autocorrelation analysis reveals strong temporal dependencies in the temperature series, with correlation remaining high even at larger lags. A clear oscillatory pattern is observed, with a peak around lag 144, confirming the presence of daily periodicity. Despite short-term fluctuations, the data exhibits quasi-periodic behavior, making it suitable for sequence modeling approaches such as RNN, LSTM, and GRU.
