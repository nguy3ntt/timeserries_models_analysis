## Overview

This dataset contains high-frequency weather observations recorded at 10-minute intervals. Preprocessing focuses on ensuring temporal consistency, handling missing values, and transforming the data into a format suitable for time series modelling.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler


df = pd.read_csv("C:/Users/ADMIN/Downloads/timeserries_models_analysis/data/raw/cleaned_weather.csv")

print(df.shape)
df.info()
df.describe()
```

```python
df.head(10)
```

```python
df['date'] = pd.to_datetime(df['date'])   
df = df.sort_values('date')               
df.set_index('date', inplace=True)       
df = df.resample('10min').mean() 
print(df.info())
df.head(10)
```

### Datetime Processing

The timestamp column was converted to a datetime format and set as the index. This enables proper time-based operations such as sorting, interpolation, and resampling if required.

### Temporal Consistency Check

The dataset was verified to ensure consistent 10-minute intervals between observations. Resampling was only applied if irregularities were detected, avoiding unnecessary data aggregation

```python
df = df.interpolate(method='time')
df = df.ffill()
df = df.dropna()
df = df.clip(
    lower=df.quantile(0.01),
    upper=df.quantile(0.99),
    axis=1
)

df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month

df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
print(df.info())
df.head(10)
```

### Missing Value Handling

Missing values were handled using time-aware interpolation, followed by forward filling to preserve temporal continuity. Any remaining missing values were removed to ensure model stability.

### Outlier Handling

Extreme values were clipped using quantile thresholds (1st and 99th percentiles) to reduce the impact of noise and measurement anomalies.

### Feature Engineering

Temporal features such as hour, day of week, and month were extracted to capture periodic patterns in the data.

Cyclical encoding (sine and cosine transformations) was applied to represent periodic features (e.g., hour of day) in a continuous manner, allowing models to better learn seasonal patterns.

```python
scaler = StandardScaler()

df_scaled = pd.DataFrame(
    scaler.fit_transform(df),
    index=df.index,
    columns=df.columns
)

train_size = int(len(df_scaled) * 0.8)

train = df_scaled[:train_size]
test = df_scaled[train_size:]

def create_sequences(data, window=144):  # 144 = 1 day (10-min intervals)
    X, y = [], []
    
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train.values)
X_test, y_test = create_sequences(test.values)
```

### Scaling

All features were standardised using z-score normalisation to ensure consistent feature magnitudes and improve model convergence.

### Sequence Construction

A sliding window approach was used to transform the continuous time series into supervised learning samples. Each input sequence consists of 144 time steps (representing 24 hours), used to predict the subsequent observation.
